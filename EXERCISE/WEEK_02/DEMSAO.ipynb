## **1. Tóm tắt bài toán**
Tìm số lượng phần tử của dãy con tăng dài nhất theo giá trị **y** của trục **Oy** sau phép xoay điểm theo một góc $\mathbf{alpha}$ 

## **2. Kỹ thuật áp dụng**
- Xoay trục tọa độ: Tìm tọa độ ảnh của một điểm khi xoay điểm đó quanh điểm O(0, 0) với góc xoay $\alpha$ theo công thức sau:

   x' = x*$\cos\alpha$  - y*$\sin\alpha$, 

   y' = x*$\sin\alpha$ - y*$\cos\alpha$ 
- Dãy con tăng dài nhất (Longest Increasing Subsequences)
- Chặt nhị phân để tìm kiếm phần tử  vị trí phần tử tăng dần tiếp theo cho dãy con tăng dài nhất

````python
import bisect

def INPUT():
	global n, a, b, c, d
	n = int(input())
	a, b, c, d = [int(j) for i in input().split() for j in i.split("/")]
	global y_alpha
	y_alpha = []
	for _ in range(n):
		x, y = [int(x) for x in input().split()]
		y_alpha.append((b * y - a * x, c * x - d * y))


if __name__ == '__main__':
	INPUT()
	y_alpha = sorted([(x, y) for x, y in y_alpha if x > 0 and y > 0], key = lambda x: (x[0], -x[1]))
	dp = []
	for x, y in y_alpha:
		i = bisect.bisect_left(dp, y)
		if i == len(dp):
			dp.append(y)
		else:
			dp[i] = y;

	print(len(dp));
````
